```python
from transformers import pipeline
```

```python
emotionModel = pipeline('text-classification',model='SamLowe/roberta-base-go_emotions')
```

```python
label = emotionModel(['I am not having a great day.',"i'm sorry that the order got delayed"])
label
```

## `Uploading dataset`

```python
import pandas as pd
```

```python
apple_df = pd.read_csv('/content/drive/MyDrive/Apple_products.csv')
apple_df.head()
```

```python
reviews = apple_df['text'].tolist()
reviews
```

```python
def getEmotion(text):
  label = emotionModel(text)
  return label[0]['label']
```

```python
reviews_emotion = [getEmotion(text) for text in reviews]
reviews_emotion
```

```python
apple_df['Emotion'] = reviews_emotion
apple_df.head()
```

```python
apple_df.iloc[100]['text']
```

```python
import plotly.express as px

# Create a countplot using Plotly Express histogram
fig = px.histogram(apple_df, x='Emotion', title='RoBertaEmotions')
fig.update_layout(barmode='stack')  # Set barmode to 'count' for countplot effect

fig.show()
```

```python
# Filter out negative emotions
negative_reviews = apple_df[apple_df['Emotion'].isin(['diapproval','annoyance','confusion','fear','disgust','embarrassement'])]
negative_reviews.head()
```
